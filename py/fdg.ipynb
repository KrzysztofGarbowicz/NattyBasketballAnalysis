```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import json
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/2019-20_pbp.csv').dropna(subset=['EnterGame', 'LeaveGame'])
```

```python
teams = pd.unique(df['HomeTeam'])
```

```python
Nodes = dict()
Links = dict()
```

```python
threshold = 0
```

```python
for team in teams:
  nodes = []
  links = []

  teamdf = df.query('HomeTeam == @team or AwayTeam == @team')
  for i in range(teamdf.shape[0]):
    sub = teamdf.iloc[i]
    p1 = sub['EnterGame']
    p2 = sub['LeaveGame']
    if (sub['HomeTeam']==team and pd.notna(sub['HomePlay'])):
      player1 = {'id': sub['EnterGame'], 'group': team, 'name':sub['EnterGame'].split(' - ')[0], 'in': df.query('EnterGame==@p1').shape[0], 'out': df.query('LeaveGame==@p1').shape[0]}
      player2 = {'id': sub['LeaveGame'], 'group': team, 'name':sub['LeaveGame'].split(' - ')[0], 'in': df.query('EnterGame==@p2').shape[0], 'out': df.query('LeaveGame==@p2').shape[0]}
    elif (sub['AwayTeam']==team and pd.notna(sub['AwayPlay'])):
      player1 = {'id': sub['EnterGame'], 'group': team, 'name':sub['EnterGame'].split(' - ')[0], 'in': df.query('EnterGame==@p1').shape[0], 'out': df.query('LeaveGame==@p1').shape[0]}
      player2 = {'id': sub['LeaveGame'], 'group': team, 'name':sub['LeaveGame'].split(' - ')[0], 'in': df.query('EnterGame==@p2').shape[0], 'out': df.query('LeaveGame==@p2').shape[0]}

    if player1 not in nodes:
        nodes.append(player1)
    if player2 not in nodes:
        nodes.append(player2)

    if player1['id']>player2['id']:
      p1 = player1['id']
      p2 = player2['id']
    else:
      p1 = player2['id']
      p2 = player1['id']
      
    subcount = df.query('EnterGame == @p1 and LeaveGame == @p2').shape[0]
    subcount += df.query('EnterGame == @p2 and LeaveGame == @p1').shape[0]
    
    if subcount>threshold:
      link = {'source': p1, 'target': p2, 'strength': subcount}
      if link not in links:
        links.append(link)

  Nodes[team]=nodes
  Links[team]=links
  
```

```python
all = dict()
all['nodes']=Nodes
all['links']=Links
```
