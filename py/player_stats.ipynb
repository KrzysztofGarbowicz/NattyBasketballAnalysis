```python
def length_conver(value):
    foot = int(value.split('-')[0])
    inch = float(value.split('-')[1])
    return round(foot*30.48+inch*2.54, 2)

def lbs_conver(value):
    return round(float(value)*0.453592, 2)
```

```python
array_Ht = df_player['Ht'].values
array_Wt = df_player['Wt'].values
for i in range(len(array_Wt)):
#     array_Ht[i] = length_conver(array_Ht[i])
    array_Wt[i] = lbs_conver(array_Wt[i])

# array_Wt
```

```python
df_out = df_player[['Pos', 'Ht', 'Wt']]
print(np.min(df_out['Ht'].values), np.max(df_out['Ht'].values), \
     np.min(df_out['Wt'].values), np.max(df_out['Wt'].values))
```

```python
pos_dict = {}
for i in df_out['Pos']:
    if i not in pos_dict:
        pos_dict[i]=1
    else:
        pos_dict[i]+=1
```
