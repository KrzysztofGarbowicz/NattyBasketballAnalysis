```python
df_pt = pd.DataFrame(columns=['Name', 'Team'])
for team in df['team'].values:
    pagestr = 'https://www.basketball-reference.com/teams/'+team+'/2020.html'
    df_ = pd.read_html(pagestr)[0]
    for name in df_['Player'].values:
        df_pt = df_pt.append({'Name':name, 'Team':team}, ignore_index=True)
```

```python
aggregation_functions = {'Pos': 'first', 'Team': 'sum', 'Name': 'first', \
                        'Ht': 'first', 'Wt': 'first'}
df_player_new = df_player_.groupby('id', as_index=False).\
aggregate(aggregation_functions).reindex(columns=df_player_.columns)
```

```python
def length_conver(value):
    foot = int(value.split('-')[0])
    inch = float(value.split('-')[1])
    return round(foot*30.48+inch*2.54, 2)

def lbs_conver(value):
    return round(float(value)*0.453592, 2)
```

```python
array_Ht = df_player_new['Ht'].values
array_Wt = df_player_new['Wt'].values
for i in range(len(array_Wt)):
    array_Ht[i] = length_conver(array_Ht[i])
    array_Wt[i] = lbs_conver(array_Wt[i])
```
