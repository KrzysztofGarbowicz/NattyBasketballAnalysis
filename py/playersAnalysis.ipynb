```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/player_data_ori.csv')
# data

data_f = data.drop(data[data.To<2000].index)
# data_f

data_f.drop('Colleges', axis=1, inplace=True)

player_id = []
player_name = []
for i in data_f['Player'].values:
    pair = i.split('\\')
    player_name.append(pair[0])
    player_id.append(pair[1])
# player_id

# data_f.insert(0, 'id', player_id, True)
data_f.insert(1, 'Name', player_name, True)
data_f.drop('Player', axis=1, inplace=True)

data_f.index = player_id

# data_f['NBA'] = 0
data_f
```

```python
data_pbp = pd.read_csv('data/2019-20_pbp.csv')
data_pbp
```

```python
data_pbp.columns
```

```python
shooters = data_pbp['Shooter'].values
assisters = data_pbp['Assister'].values
blockers = data_pbp['Blocker'].values
foulers = data_pbp['Fouler'].values
fouleds = data_pbp['Fouled'].values
rebounders = data_pbp['Rebounder'].values
ft_shooters = data_pbp['FreeThrowShooter'].values
enters = data_pbp['EnterGame'].values
leavers = data_pbp['LeaveGame'].values
t_players = data_pbp['TurnoverPlayer'].values
t_causers = data_pbp['TurnoverCauser'].values
j_a_players = data_pbp['JumpballAwayPlayer'].values
j_h_players = data_pbp['JumpballHomePlayer'].values
j_posses = data_pbp['JumpballPoss'].values
```

```python
player_pbp_set = set()
for i in [shooters, assisters, blockers, foulers, fouleds, rebounders, ft_shooters, enters, leavers, t_players, t_causers, j_a_players, j_h_players, j_posses]:
    for j in i:
        if not pd.isnull(j):
            name_code = j.split()[-1]
            if not (name_code[-1] == 'c' or name_code == 'Team'):
                    player_pbp_set.add(name_code)
```

```python
print(player_pbp_set)
```

To check if there are missed players:

```python
for i in player_pbp_set:
    if not i in data_f.index:
        print(i)
```

Remove the players not showed in 2019-20_pbp.csv:

```python
for i in data_f.index:
    if i not in player_pbp_set:
        data_f.drop([i], inplace = True)
```

Calculate the age of the players:

```python
from datetime import date
 
def calculateAge(birthDate):
    today = date.today()
    age = today.year-birthDate.year-((today.month, today.day)<(birthDate.month, birthDate.day))
    return age
     
# Driver code
# print(calculateAge(date(1997, 2, 3)), "years")
```

```python
ages = []
for i in data_f['Birth Date'].values:
    if not pd.isnull(i):
        day = i.split('/')
        ages.append(calculateAge(date(int(day[-1]), int(day[0]), int(day[1]))))
```

```python
data_f.insert(7, 'Age', ages, True)
```

```python
df_inten = pd.read_csv('data/intensity.csv')
df_player = pd.read_csv('data/players.csv')
```

```python
df_player
```

Find the player-team pair form the tables on https://www.basketball-reference.com/teams/ 

```python
df_pt = pd.DataFrame(columns=['Name', 'Team'])
for team in df_inten['team'].values:
    pagestr = 'https://www.basketball-reference.com/teams/'+team+'/2020.html'
    df_ = pd.read_html(pagestr)[0]
    for name in df_['Player'].values:
        df_pt = df_pt.append({'Name':name, 'Team':team}, ignore_index=True)
```

```python
display(df_pt)
```

```python
df_player_ = pd.merge(df_player, df_pt, how='outer')
df_player_ = df_player_.drop(columns=['Birth Date', 'From', 'To', 'Age'])
df_player_ = df_player_.rename(columns={'Unnamed: 0':'id'})
df_player_ = df_player_[df_player_.Pos.notnull()]
```

```python
display(df_player_)
```

Some players served in two teams in the season.

```python
aggregation_functions = {'Pos': 'first', 'Team': 'sum', 'Name': 'first', \
                        'Ht': 'first', 'Wt': 'first'}
df_player_new = df_player_.groupby('id', as_index=False).\
aggregate(aggregation_functions).reindex(columns=df_player_.columns)
```

Add '/' between two teams:

```python
team_arr = df_player_new['Team'].values
for i in range(len(team_arr)):
    if len(team_arr[i])>3:
        team_arr[i] = team_arr[i][:3]+'/'+team_arr[i][3:]
```

Convert the imperial units to metric units:

```python
def length_conver(value):
    foot = int(value.split('-')[0])
    inch = float(value.split('-')[1])
    return round(foot*30.48+inch*2.54, 2)

def lbs_conver(value):
    return round(float(value)*0.453592, 2)
```

```python
array_Ht = df_player_new['Ht'].values
array_Wt = df_player_new['Wt'].values
for i in range(len(array_Wt)):
    array_Ht[i] = length_conver(array_Ht[i])
    array_Wt[i] = lbs_conver(array_Wt[i])
```
