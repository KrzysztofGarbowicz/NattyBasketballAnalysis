```python
intensity_dict = {}
for i in range(1, 1144):
    df_match = pd.read_csv('data/matches/'+str(i)+'.csv')
    team_away = df_match.loc[0]['AwayTeam']
    team_home = df_match.loc[0]['HomeTeam']
    if team_away not in intensity_dict:
        intensity_dict[team_away]={'attempt': 0, 'make': 0, 'miss': 0, 'sum': 0,\
                                  'attempt_q1': 0, 'make_q1': 0, 'miss_q1': 0,\
                                  'attempt_q2': 0, 'make_q2': 0, 'miss_q2': 0,\
                                  'attempt_q3': 0, 'make_q3': 0, 'miss_q3': 0,\
                                  'attempt_q4': 0, 'make_q4': 0, 'miss_q4': 0,\
                                  'attempt_q5': 0, 'make_q5': 0, 'miss_q5': 0,\
                                  'attempt_q6': 0, 'make_q6': 0, 'miss_q6': 0}
    if team_home not in intensity_dict:
        intensity_dict[team_home]={'attempt': 0, 'make': 0, 'miss': 0, 'sum': 0,\
                                  'attempt_q1': 0, 'make_q1': 0, 'miss_q1': 0,\
                                  'attempt_q2': 0, 'make_q2': 0, 'miss_q2': 0,\
                                  'attempt_q3': 0, 'make_q3': 0, 'miss_q3': 0,\
                                  'attempt_q4': 0, 'make_q4': 0, 'miss_q4': 0,\
                                  'attempt_q5': 0, 'make_q5': 0, 'miss_q5': 0,\
                                  'attempt_q6': 0, 'make_q6': 0, 'miss_q6': 0}
    
    intensity_dict[team_away]['sum']+=1
    intensity_dict[team_home]['sum']+=1
    
    for index, row in df_match.iterrows():
        if pd.isnull(row['ShotOutcome']):
            continue
        if pd.isnull(row['AwayPlay']) == False:
            intensity_dict[team_away]['attempt']+=1
            aqstr = 'attempt_q'+str(row['Quarter'])
            intensity_dict[team_away][aqstr]+=1
            if row['ShotOutcome'] == 'make':
                intensity_dict[team_away]['make']+=1
                mqstr = 'make_q'+str(row['Quarter'])
                intensity_dict[team_away][mqstr]+=1
                continue
            else:
                intensity_dict[team_away]['miss']+=1
                mqstr = 'miss_q'+str(row['Quarter'])
                intensity_dict[team_away][mqstr]+=1
                continue
        else:
            intensity_dict[team_home]['attempt']+=1
            aqstr = 'attempt_q'+str(row['Quarter'])
            intensity_dict[team_home][aqstr]+=1
            if row['ShotOutcome'] == 'make':
                intensity_dict[team_home]['make']+=1
                mqstr = 'make_q'+str(row['Quarter'])
                intensity_dict[team_home][mqstr]+=1
                continue
            else:
                intensity_dict[team_home]['miss']+=1
                mqstr = 'miss_q'+str(row['Quarter'])
                intensity_dict[team_home][mqstr]+=1
                continue
```
